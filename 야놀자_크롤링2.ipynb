```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import json
from tqdm import tqdm
```

```python
yanolja=pd.read_csv("yanolja.csv")
yanolja["comments"]=np.nan
yanolja.head()
```

```python
#for i in range(yanolja.shape[0]): 300,
for i in tqdm(range(yanolja.shape[0])):
    id=yanolja["url"][i].split('/')[-1]
    url=' https://place-site.yanolja.com/api/stay.properties.review-list.get,stay.properties.review-top.get?batch=1&input=%7B%220%22%3A%7B%22propertyId%22%3A{}%2C%22query%22%3A%7B%22photoReviewOnly%22%3Afalse%2C%22roomTypeIds%22%3A%22%22%2C%22page%22%3A1%2C%22sort%22%3A%22BEST%22%7D%7D%2C%221%22%3A%7B%22propertyId%22%3A{}%2C%22query%22%3A%7B%22roomTypeIds%22%3A%22%22%7D%7D%7D'.format(id,id)
    resp=requests.get(url)
    jsn=json.loads(resp.content)
    jsn=jsn[0]
    review_list=jsn["result"]['data']['reviews']

    my_reviews=[]
    my_sores=[]
    #for review in review_list:
    #    my_reviews.append(' '.join(review['content']))
    #yanolja["comments"][i]=my_reviews
    try:
        for review in review_list:
            my_reviews.append(' '.join(review['content']))
            #'score'
            my_scores.append(' '.join(review['score']))
        yanolja["comments"][i]=my_reviews
    except:
        continue
```

```python
for i in tqdm(range(yanolja.shape[0])):
    yanolja['review_num'][i]=len(yanolja['comments'][i])
```

```python
#저장
yanolja.to_csv("yanolja_comments.csv", index=False, encoding="utf-8-sig")
```
