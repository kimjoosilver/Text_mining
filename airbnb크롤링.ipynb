```python
from tqdm.notebook import tqdm
```

```python
title = []
star = []
url = []
facility = []
reviews = [] 
review_cnt= []
last_num= 0 
temp_list = [] 
soup = bs(driver.page_source, 'html.parser')
temp = soup.select('div._jro6t0 a')
for i in temp:
    if i.text != '':
        temp_list.append(i.text)

#맨 마지막 페이지 가져오는 방법 
for i in temp_list:
    last_num = max(last_num,int(i))


for i in tqdm(range(last_num)):
    temp1 = soup.select("div.g1qv1ctd.cb4nyux.dir.dir-ltr")
    #한페이지에 제목 평점, 리뷰 개수 가져오기 
    for i in temp1:
        try:
            title.append(i.select('span.t6mzqp7.dir.dir-ltr')[0].text)
            star.append(i.select('span.r1dxllyb.dir.dir-ltr')[0].text.split(' ')[0])
            review_cnt.append(int(i.select('span.r1dxllyb.dir.dir-ltr')[0].text.split(' ')[1].replace('(', '').replace(')', '')))
        except:
            review_cnt.append(0)
    #url 가져오기 
    temp2 = soup.select('a.bn2bl2p.dir.dir-ltr')
    for i in tqdm(temp2):
        temp_url = "https://www.airbnb.co.kr" + i['href']
        url.append(temp_url)    
        
    
    #한페이지 url 들어가기 
        driver.get(temp_url)
        driver.implicitly_wait(10)
        
        driver.find_element(By.CLASS_NAME, 'b65jmrv.v7aged4.dir.dir-ltr').click()
        temp_facility = []
        #편의 용품창 켜서 가져오기 
        for i in driver.find_elements(By.CLASS_NAME, '_11jhslp'):
            temp_plus = []
            temp_plus.append(i.text.replace('\n', ' '))
            for j in temp_plus:
                temp_facility.append(j)
        facility.append(temp_facility)
        #편의 용품창 끄기 
        driver.implicitly_wait(10)
        driver.find_element(By.CLASS_NAME, '_pa35zs').click()
        driver.implicitly_wait(10)
        #리뷰창 키기 
        driver.find_element(By.XPATH, '/html/body/div[5]/div/div/div[1]/div/div[1]/div/div/div/div[1]/main/div/div[1]/div[4]/div/div/div/div[2]/section/div[4]/button').click()

        
        #리뷰용 자동 스크롤 넘버 거지고 오기  리뷰 스크롤당 7개니까 리뷰 개수 / 7 해서 +1 한다. 
        temp_num = driver.find_element(By.XPATH, '/html/body/div[9]/section/div/div/div[2]/div/div[3]/div/div/div/section/div[1]/div[1]/h2').text
        scroll_num = temp_num.split(' ')[-1]
        scroll_num = int(scroll_num.replace('개', ''))
        scroll_num = scroll_num // 7 + 1
        scroll_num

        all_review = driver.find_element(By.CLASS_NAME, '_17itzz4')
        for i in range(scroll_num):
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', all_review)
            driver.implicitly_wait(10)
        review = []
        for i in driver.find_elements(By.CLASS_NAME, 'r1are2x1.dir.dir-ltr'):
            review.append(i.text)

        reviews.append(review) 
        driver.implicitly_wait(10)
    
        for i in range(2):
            driver.back()
            driver.implicitly_wait(10)
        
    #다음 페이지로 
    driver.find_element(By.XPATH, '/html/body/div[5]/div/div/div[1]/div/div[1]/div/div/div/div/div/div[2]/main/div[2]/div[3]/div/div/nav/div/a[5]').click()
    soup = bs(driver.page_source, 'html.parser')
    driver.implicitly_wait(10)
```
